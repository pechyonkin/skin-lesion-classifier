```python
df2 = df.copy()
```

```python
new_labels = {
    'pigmented benign keratosis': 'benign', 
    'basal cell carcinoma': 'malignant',
    'actinic keratosis': 'indeterminate/malignant', 
    'vascular lesion': 'indeterminate', 
    'dermatofibroma': 'benign',
    'squamous cell carcinoma': 'malignant',
}
```

```python
for key, value in new_labels.items():
    mask_null = df2['benign_malignant'].isnull()
    mask_key = df['diagnosis'] == key
    df2.loc[mask_null & mask_key, 'benign_malignant'] = value
```

```python
mask_null = df2['benign_malignant'].isnull()
mask_pgk = df['diagnosis'] == 'pigmented benign keratosis'
df2.loc[mask_null & mask_pgk, 'benign_malignant'] = 'benign'
```

```python
df2[df2['benign_malignant'].isnull()]['diagnosis'].unique()
```

```python
idx_to_drop = df2[df2['benign_malignant'].isnull()].index
```

```python
df3 = df2.drop(idx_to_drop)
```

```python
df3[df3['benign_malignant'].isnull()]['diagnosis'].unique()
```

```python
df3['benign_malignant'].value_counts()
```

```python
df3.loc[df3['benign_malignant'] == 'indeterminate/malignant', 'benign_malignant'] = 'malignant'
```

```python
df3['benign_malignant'].value_counts()
```

```python
mask_ind = df3['benign_malignant'] == 'indeterminate'
mask_ind_ben = df3['benign_malignant'] == 'indeterminate/benign'
idx_ind_to_drop = df3[mask_ind | mask_ind_ben].index
df4 = df3.drop(idx_ind_to_drop)
```
