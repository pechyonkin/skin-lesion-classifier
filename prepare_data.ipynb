```python
from pathlib import Path
import pathlib
import json
import pandas as pd
```

```python
PATH = Path("/Users/fazzl/data/isic")
```

```python
def hidden(p):
    'is the path p hidden?'
    if type(p) == str:
        p = Path(p)
    return str(p.parts[-1])[0] == '.'

def ls_files(p):
    return [e for e in p.iterdir() if not hidden(e)]
```

```python
descrs = ls_files(PATH / 'descriptions')
```

```python
t = descrs[0]
t = str(t) # need to be string for json to open
```

```python
def make_pandas(descriptions=descrs, limit=100):
    """ Convert ISIC individual json descriptions into a pandas dataframe"""
    
    target_keys_top = [
    '_id', 
    'name',
    ]

    target_keys_clinical = [
        'diagnosis', 
        'benign_malignant', 
        'melanocytic', 
        'age_approx', 
        'sex', 
    ]

    target_keys_acquisition = [
        'image_type', 
        'pixelsX', 
        'pixelsY', 
    ]
    
    descriptions = descriptions[:limit] if limit is not None else descriptions
    rows_list = []
    for d in descriptions:
        with open(d) as f:
            data = json.load(f)
        row = {}
        for k in target_keys_top:
            row[k] = data[k] if k in data.keys() else None
        for k in target_keys_clinical:
            row[k] = data['meta']['clinical'][k] if k in data['meta']['clinical'].keys() else None
        for k in target_keys_acquisition:
            row[k] = data['meta']['acquisition'][k] if k in data['meta']['acquisition'].keys() else None
        rows_list.append(row)
    return pd.DataFrame(rows_list)
```

```python
df = make_pandas(descrs, limit=None)
```
